##Amazon Product Review Dataset

Product Review Dataset Description

This dataset contains comprehensive information about various products and their corresponding reviews. The dataset covers a range of products from different categories, capturing insights from customer reviews, ratings, and related attributes. The dataset is structured with the following columns:

Product_id: Unique identifier for each product.
Product_description: Textual description providing details about the product's features, specifications, and functionality.
Product_name: The name or title of the product.
category: The category or type of the product, enabling categorization into distinct groups.
discount_percent: The percentage of discount offered on the product.
rating: Numerical rating assigned by customers to express their satisfaction with the product.
rating_count: The count of ratings received by the product, reflecting its popularity.
about_product: Additional information or summary about the product.
user_id: Unique identifier for each user providing a review.
user_name: The name associated with the user.
review_id: Unique identifier for each review.
review_title: A concise title that summarizes the content of the review.
review_content: Detailed content of the user's review, providing insights into their experiences and opinions.
img_link: Links to images associated with the product.
product_link: URLs directing to the product's webpage for further information and purchase.


The dataset is rich with information, offering opportunities for sentiment analysis, user behavior patterns, product popularity analysis, and more. With attributes ranging from user-generated content to product specifications, this dataset provides a comprehensive view of customer interactions with various products, making it valuable for both consumer insights and business decision-making.

```python
#required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importing dataset
df= pd.read_csv('amazon_review.csv')
df
```

```python
df.info()
```

```python
# Convert columns to desired data types

# Convert discount_percent to numeric
df['discount_percent'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Convert rating to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Convert rating_count to numeric
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Display basic statistics and structure of the modified dataset
print(df.info())
```

```python
# Calculate basic statistics
average_rating = df['rating'].mean()
highest_rated_product = df[df['rating'] == df['rating'].max()]['product_name'].iloc[0]
print(average_rating)
```

```python
# Histogram of ratings
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='rating', bins=5, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
#rightly skewed rating
```

```python
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split

# Load the dataset
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df[['user_id', 'product_id', 'rating']], reader)

# Split the data into train and test sets
trainset, testset = train_test_split(data, test_size=0.2)

# Create and train the model (user-based collaborative filtering)
sim_options = {
    'name': 'cosine',
    'user_based': True
}
model = KNNBasic(sim_options=sim_options)
model.fit(trainset)

# Get recommendations for a user
user_id = 'your_user_id_here'
user_products = df[df['user_id'] == user_id]['product_id'].values
user_unrated_products = [prod_id for prod_id in df['product_id'].unique() if prod_id not in user_products]

user_recs = []
for prod_id in user_unrated_products:
    predicted_rating = model.predict(user_id, prod_id).est
    user_recs.append((prod_id, predicted_rating))

# Sort recommendations by predicted rating
user_recs.sort(key=lambda x: x[1], reverse=True)

# Print top N recommended products
top_n = 5
for rec in user_recs[:top_n]:
    product_name = df[df['product_id'] == rec[0]]['product_name'].iloc[0]
    print(f"Product: {product_name}, Predicted Rating: {rec[1]}")
```

```python
from urllib.parse import urlparse

# Function to extract information from a product link
def analyze_product_link(link):
    parsed_url = urlparse(link)
    domain = parsed_url.netloc
    path_segments = parsed_url.path.strip('/').split('/')

    # Extract relevant information
    platform = domain
    category = None
    product_id = None

    if len(path_segments) >= 2:
        category = path_segments[-2]

    if len(path_segments) >= 1:
        product_id = path_segments[-1]

    return {
        'link': link,
        'platform': platform,
        'category': category,
        'product_id': product_id
    }

# Apply the link analysis function to each product link in the 'product_link' column
for index, row in df.iterrows():
    product_link = row['product_link']
    analysis_result = analyze_product_link(product_link)

    # Update the DataFrame with the analysis result
    df.at[index, 'platform'] = analysis_result['platform']
    df.at[index, 'category'] = analysis_result['category']
    df.at[index, 'product_id'] = analysis_result['product_id']

# Display the updated DataFrame
print(df[['product_link', 'platform', 'category', 'product_id']])
```
